## **Oscilador Harmônico 1D**

Importando as bibliotecas básicas

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Referencial Teórico:**
No caso de oscilador em uma dimenção, de frenquência angular $\omega$
$$
m\frac{d^{2}x}{dx^{2}} = -kx
$$
$$
\dot{x} + \omega^{2}x = 0 
$$ 

onde $\omega$ 
$$
\omega^{2}= \frac{k}{m}
$$

Vamos criar um simulação para nosso oscilador. Podemos considerar um sitema massa mola que oscile em uma direção em um espaço de apenas 1m e que tenha uma oscilaçao por segundo. 

```python
A = 1 # Amplitude de oscilação, [A] = [m]
w =1  # Frequência/velocidade angular, [w] = [1rad/s]
fase = 0 # fase inicial, representa a posiçao de onde começamos o movimento   
```

Agora precismaos criar um série de pontos para o parâmetro de tempo, vamos utilizar a função [**``np.lispace()``**](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html) da biblioteca [**``numpy``**](https://numpy.org/doc/stable/index.html)

Essa função recebe três entradas, um valor inicial, um valor final e por ultimo o número de ponto que queremos entre os dois primeiros. 

```python
t = np.linspace(0,10,100)
```

Utilizando $t$ na solução  $x(t)$ geramos um série de pontos que fazem parte da solução  

```python
x = A*np.cos(w*t+fase)
```

O que otivemos na realidade é uma série de pares finits de pontos

```python
plt.scatter(t,x) # Pontos otidos
plt.plot(t,x)    # Liga os pontos 
plt.xlabel("Tempo") 
plt.ylabel("Posição")
plt.show()
```

## **Entendendo a tragétoria**

Podemos simplificar utilizando $(\alpha - \beta)  = \delta$ que representa a diferença de fase entre $x(t)$ e $y(t)$

$
x(t) = Acos(\omega t) \\ y(t) = Bcos(\omega t+ \delta)
$

$$
y(t) = B \cos(\omega_0 t  + \delta) \\
y(t) = B \cos(\omega_0 t )\cos(\delta) - B \sin(\omega_0 t )\sin(\delta)\\
$$

 Vamos utilizar o fato de $cos(\omega t) = \frac{x(t)}{A}$

$$
y(t) = B \frac{x(t)}{A} \cos(\delta) - B \sin(\omega_0 t)\sin(\delta)
$$

Utilizando a relação trigonométrica fundamental, obtemos: 
$$
sin(\omega t)+ = \sqrt{1 -  \frac{x^{2}}{A^{2}}}
$$

$$
A y - B x \cos\delta = -B \sqrt{A^2 - x^2} \, \sin\delta
$$

Elevando ao quadrado, essa equação se torna

$$
A^2y^2 - 2ABxy \cos \delta + B^2x^2 \cos^2 \delta = A^2B^2 \sin^2 \delta - B^2x^2
$$

de modo que

$$
B^2x^2 - 2ABxy \cos \delta + A^2y^2 = A^2B^2 \sin^2 \delta
$$

Se $\delta \in \mathbb{R}$ definido como $\pm \pi/2$, essa equação fica reduzida à equação facilmente reconhecida de uma elipse:

$$\frac{x^2}{A^2} + \frac{y^2}{B^2} = 1$$


Se as amplitudes forem iguais, $A = B$ e, se $\delta = \pm \pi/2$, temos o caso especial de movimento circular:

$$x^2 + y^2 = A^2, \quad \text{para } A = B \text{ e } \delta = \pm \pi/2$$

```python
A = B  = 1 # [m]
w1 = w2 = 1 # [rad/s]
fase = np.pi/2
```

```python
x = A*np.cos(w1*t)
y = B*np.cos(w2*t+fase)
```

```python
plt.plot(x,y)
plt.title(rf'$\delta={int(np.degrees(fase))}$')
plt.xlabel("X") 
plt.ylabel("Y")
plt.show()
```

Outro caso especial resulta no caso de anulação da fase $\delta$; então, temos

$$B^2 x^2 - 2ABxy + A^2 y^2 = 0, \quad \delta = 0$$

Fatorando,

$$(Bx - Ay)^2 = 0$$


que é a equação de uma linha reta:

$$y = \frac{B}{A}x, \quad \delta = 0$$

```python
A = B  = 1 # [m]
w1 = w2 = 1 # [rad/s]
fase = 0
```

```python
x = A*np.cos(w1*t)
y = B*np.cos(w2*t+fase)
```

```python
plt.plot(x,y)
plt.title(rf'$\delta={int(np.degrees(fase))}$')
plt.xlabel("X") 
plt.ylabel("Y")
plt.show()
```

De modo similar, a fase $\delta = \pm \pi$ produz a linha reta com inclinação oposta:

$$y = -\frac{B}{A}x, \quad \delta = \pm \pi$$

```python
A = B  = 1 # [m]
w1 = w2 = 1 # [rad/s]
fase = np.pi
```

```python
x = A*np.cos(w1*t)
y = B*np.cos(w2*t+fase)
```

```python
plt.plot(x,y)
plt.title(rf'$\delta={int(np.degrees(fase))}$')
plt.xlabel("X") 
plt.ylabel("Y")
plt.show()
```

## **Oscilador isotrópico**

Em um oscilador isotrópico as frenquências angulares $\omega$ associadas a oscilação a cada direção são iguais, $\omega_{x} = \omega_{z} = \omega_{z} $. Vamos fazer plots para diderentes $\delta$, por exemplo de $0$ até $360$.


```python
fases = np.linspace(0,360,9)
fases
```

vamos determinar as constantes para $x(t)$ com índice 1 e $y(t)$ com índice 2. Para facilitar e deixar o código mais limpo, vamos colocar essas contantes em um lista.

```python
constantes = [w1, w2, A, B]
```

vamos criar um função para limpar o código. Ela recebe a lista de constantes, a lista de fases em graus e os pontos $t$, como isso ela retorna os pontos $x$ e $y$. A outra função é responsavél apenas pelos plots e não retorna variáveis.

```python
def pontos(dif_fase, t , constantes):

    fase_rad = np.radians(dif_fase)     # importante para passar valores em rad pra as funções
    fase_deg = np.round(np.degrees(fase_rad),0) # importante para legenda
    
    # Descompactando a lista de constantes
    w1, w2, A, B = constantes[0], constantes[1],constantes[2],constantes[3]

    x = A*np.cos(w1*t)
    y = B*np.cos(w2*t - fase_rad)

    return x, y, fase_deg 
```

```python
def plot(fase_val, t, constantes):
    for i, fase_val in enumerate(fases, 1):
        x, y, fase = pontos(fase_val, t, constantes)
        plt.subplot(3, 3, i)
        plt.plot(x, y)
        plt.title(rf'$\delta={int(fase)}$')  
```

```python
constantes = [1, 1, 1, 1]
plot(fase, t, constantes)

plt.tight_layout()
plt.show()
```

Se as frequências angulares $\omega$ são diferentes, por exemplo $\omega_{1} = 2\omega_{2}$

```python
constantes = [1, 2, 1, 1]
plot(fase, t, constantes)

plt.tight_layout()
plt.show()
```

```python
constantes = [3, 2, 1, 1]
plot(fase, t, constantes)

plt.tight_layout()
plt.show()
```

```python
constantes = [3, 5, 1, 1]
plot(fase, t, constantes)

plt.tight_layout()
plt.show()
```

Em um oscilador isotrópico o movimento é periódico, as tragetória é fechada. Isso acontece quando $\frac{\omega_{1}}{\omega_{2}}$ é um número racional. Quando a razão é um número irracional, o movimento não é periódico e a curva não se fecha.  Chamamos esse caso de oslidador Anisotropico.

## **Oscilador Anisotropico** 

Podemos ter por $\omega_{1}=e$, $\omega_{1}=\pi$

```python
constantes = [np.e, np.pi, 1, 1]
plot(fase, t, constantes)
plt.tight_layout()
plt.show()
```

## **Rastro**

Vamos o obter o movimento do ponto oscilante e rastro de sua tragétoria

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Parâmetros
A,B = 1,1
omega_x, omega_y = 1, np.exp(1)  #rad/s
d_fase = np.radians(30)
N_voltas = 10
t = np.arange(0.0, N_voltas*(2*np.pi), 0.05)  # len(t) quantidade de frames
x_points = A*np.cos(omega_x*t)  # Função x(t)
y_points = B*np.sin(omega_y*t+d_fase)  # Função y(t)

# Configurar a figura e o eixo
fig, ax = plt.subplots()
ax.set_xlim(-A-0.5, A+0.5)
ax.set_ylim(-B-0.5, B+0.5)
ax.set_aspect('equal')  # Manter proporção 1:1
ax.grid(True)

# Inicializar linha vazia
line, = ax.plot([], [], 'ro-', markersize=8)  # 'ro-' significa pontos vermelhos com linha
trail, = ax.plot([], [], 'b-', alpha=0.3)  # Trilha azul transparente

# Listas para armazenar a trilha
x_trail = []
y_trail = []

# Função de inicialização
def init():
    line.set_data([], [])
    trail.set_data([], [])
    return line, trail

# Função de atualização para cada frame
def update(frame):
    line.set_data([x_points[frame]], [y_points[frame]])
    
    # Adicionar ponto à trilha
    x_trail.append(x_points[frame])
    y_trail.append(y_points[frame])
    
    # Atualizar trilha
    trail.set_data(x_trail, y_trail)
    
    return line, trail

# Criar a animação
ani = FuncAnimation(fig,
                    update,
                    frames=len(t),
                    init_func=init,
                    interval=50,   # 5 ms/frame
                    blit=True)

# Exibir a animação
HTML(ani.to_html5_video())
```
